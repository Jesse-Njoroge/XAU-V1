```python
import pandas as pd
from pathlib import Path

def partition_training_data(input_path, output_folder):
    """
    Splits the labeled data into 3 files and filters for specific training features.
    """
    
    # 1. Load the labeled dataset
    print(f"Loading labeled data from: {input_path}")
    df = pd.read_csv(input_path, index_col=0, parse_dates=True)
    
    # 2. Define the specific columns required for training
    training_columns = [
        'XAU_Returns', 
        'DXY_Returns', 
        'SPX_Returns', 
        'US10Y_Diff', 
        'VIX_Diff', 
        'Target_Next_Day_Gold'
    ]

    # Check if 'Regime' exists (needed for splitting) but not for final training CSVs
    if 'Regime' not in df.columns:
        print("Error: 'Regime' column not found.")
        return

    # 3. Define the regimes and filenames
    regimes = {
        0: '0_LOW_VOL_TRAINING.csv',
        1: '1_MED_VOL_TRAINING.csv',
        2: '2_HIGH_VOL_TRAINING.csv'
    }

    print("Filtering and Partitioning data...")

    for r_id, filename in regimes.items():
        # A. Filter for the specific regime
        regime_df = df[df['Regime'] == r_id]
        
        # B. Keep ONLY the training columns requested
        # We use errors='ignore' in case a column is missing to prevent a crash
        final_training_df = regime_df[training_columns]
        
        # C. Save to the database folder
        output_path = output_folder / filename
        final_training_df.to_csv(output_path, index=True)
        
        print(f"Regime {r_id}: Saved {len(final_training_df)} rows to {filename}")

# =============================================================================
# EXECUTION
# =============================================================================
if __name__ == "__main__":
    BASE_DIR = Path.cwd() / 'database'
    INPUT_FILE = BASE_DIR / 'labeled_feature_matrix.csv'
    
    partition_training_data(INPUT_FILE, BASE_DIR)
    
    print("\n✅ Training-ready files (features only) are now in the database folder.")
```
