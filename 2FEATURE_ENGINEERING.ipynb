```python
import pandas as pd
import numpy as np
from pathlib import Path

BASE_DIR = Path.cwd()
DATABASE_DIR = BASE_DIR / 'database'

df = pd.read_csv(DATABASE_DIR / 'database.csv', index_col=0)
# --- Feature Engineering ---

# Log Returns for Prices (Percentage-based movement)
df['XAU_Returns'] = np.log(df['XAUUSD_Close'] / df['XAUUSD_Close'].shift(1))
df['DXY_Returns'] = np.log(df['DXY_Close'] / df['DXY_Close'].shift(1))
df['SPX_Returns'] = np.log(df['SPX_Close'] / df['SPX_Close'].shift(1))

# First Differences for Rates/Volatility (Point-based movement)
df['US10Y_Diff'] = df['US10Y_Close'].diff()
df['VIX_Diff'] = df['VIX_Close'].diff()

# --- The "Target" Column (Phase 2 of your Hypothesis) ---
# We want to predict TOMORROW'S gold return using TODAY'S macro data.
# We shift the gold returns BACKWARDS by 1 so the 'Target' is next day's move.
df['Target_Next_Day_Gold'] = df['XAU_Returns'].shift(-1)

# --- Cleanup ---
# Remove the first row (has NaNs from .shift(1)) 
# and the last row (has NaN in Target from .shift(-1))
df_final = df.dropna()

# --- Selecting Columns ---
# We only want to save the features and the target, not the raw prices
features_to_save = [
    'XAU_Returns', 'DXY_Returns', 'SPX_Returns', 
    'US10Y_Diff', 'VIX_Diff', 'XAUUSD_Close', 'XAUUSD_Open', 'XAUUSD_High', 'XAUUSD_Low','Target_Next_Day_Gold'
]

# --- Save to CSV ---
df_final[features_to_save].to_csv(DATABASE_DIR / 'feature_matrix.csv')

print("Feature matrix saved successfully!")
```
